```python
import pandas as pd
import requests
from config import api_key
import json
```

```python
gop_candidate_df=pd.read_csv("2020_rep_candidate.csv")
```

```python
url="https://api.open.fec.gov/v1/schedules/schedule_a/by_state/by_candidate/"
```

```python
gop_can_id=gop_candidate_df.candidate_id
gop_can_id.to_list()
```

```python
def extract_data(results):
    return {
    "candidate_id":results["candidate_id"],
    "state": results["state"],
    "donation_count": results["count"],
    "total_raised": results["total"],
    "cycle": results["cycle"]
    }
```

```python
gop_finance_3 = []
# pages=range(1,3)

for candidate in gop_can_id:
#     for page in range(1,3):

    data = requests.get(url=url, params={
        "election_full": False,
        "sort_hide_null": False,
        "cycle": 2020,
        "api_key": api_key,
        "sort_nulls_last": False,
        "sort_null_only": False,
        "per_page":  20,
        "candidate_id": candidate,
        "page": 3
        
    }).json()['results']
    
    for result in data:
        gop_finances_3 = extract_data(result)
    
        gop_finance_3.append(gop_finances_3)
```

```python
gop_finances_3=pd.DataFrame(gop_finance_3)
gop_finances_3.head()
```

```python
gop_finance_2 = []
# pages=range(1,3)

for candidate in gop_can_id:
#     for page in range(1,3):

    data = requests.get(url=url, params={
        "election_full": False,
        "sort_hide_null": False,
        "cycle": 2020,
        "api_key": api_key,
        "sort_nulls_last": False,
        "sort_null_only": False,
        "per_page":  20,
        "candidate_id": candidate,
        "page": 2
        
    }).json()['results']
    
    for result in data:
        gop_finances_2 = extract_data(result)
    
        gop_finance_2.append(gop_finances_2)
```

```python
gop_finances_2=pd.DataFrame(gop_finance_2)
gop_finances_2.head()
```

```python
gop_finance_1 = []
# pages=range(1,3)

for candidate in gop_can_id:
#     for page in range(1,3):

    data = requests.get(url=url, params={
        "election_full": False,
        "sort_hide_null": False,
        "cycle": 2020,
        "api_key": api_key,
        "sort_nulls_last": False,
        "sort_null_only": False,
        "per_page":  20,
        "candidate_id": candidate,
        "page": 1
        
    }).json()['results']
    
    for result in data:
        gop_finances_1 = extract_data(result)
    
        gop_finance_1.append(gop_finances_1)
```

```python
gop_finances_1=pd.DataFrame(gop_finance_1)
gop_finances_1.head()
```

```python
concatting=[gop_finances_1, gop_finances_2, gop_finances_3]
new_gop_finance=pd.concat(concatting)
new_gop_finance.to_csv("2020_gop_finance_data.csv")
```
