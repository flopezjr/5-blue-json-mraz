```python
import pandas as pd
import requests
from config import api_key
import json
```

```python
url="https://api.open.fec.gov/v1/candidates/"
params={
    "has_raised_funds":True,
    "sort_null_only":False,
    "election_year":2018,
    "office":"S",
    "api_key":api_key,
    "party":"Rep"
}
data=requests.get(url=url,params=params).json()
```

```python
candidate_page_1=pd.DataFrame(data["results"])
candidate_page_1.head()
```

```python
url="https://api.open.fec.gov/v1/candidates/"
params_5={
    "has_raised_funds":True,
    "sort_null_only":False,
    "election_year":2018,
    "office":"S",
    "api_key":api_key,
    "party":"Rep",
    "page":5
}
page_5=requests.get(url=url,params=params_5).json()
```

```python
candidate_page_5=pd.DataFrame(page_5["results"])
candidate_page_5.head()
```

```python
# all page dfs are added to a list so they can be combined
# into a single df via concatenation 

concat_reps = [candidate_page_1, candidate_page_2, candidate_page_3, candidate_page_4, candidate_page_5]
```

```python
concat_reps
```

```python
# new df created by concat'ing the list above
rep_candidates=pd.concat(concat_reps)
```

```python
rep_candidates.head()
```

```python
# final df for dems created by narrowing the concatenated df
# by the columns we want to keep

rep_candidates=rep_candidates[["name", "candidate_id", "state", "party", "incumbent_challenge_full", "office_full", "cycles", "election_years", "party", "candidate_status"]]
```

```python
# test the final df shape to make sure all records were
# successfully added. when i ran the postman call, i saw
# that the total list contained 81 records, so this df
# successfully contains all candidates
rep_candidates.reset_index()
```

```python
rep_candidates.to_csv("rep_candidate.csv")
```
