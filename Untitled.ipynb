```python
import pandas as pd
```

```python
state_status=pd.read_csv("state_status.csv")
```

```python
state_status.head()
```

```python
candidates=pd.read_csv("2020_final_cand")
```

```python
status=pd.DataFrame.merge(candidates,state_status, left_on="candidate_state", right_on="State")
```

```python
status
```

```python
can_status=status["party"]==status["status"]
```

```python
can_status=can_status.to_frame()
```

```python
candidates_2020=pd.DataFrame.merge(candidates,can_status, left_index=True,right_index=True)
```
